### INTRODUCERE
Aceasta analiza exploreaza modul in care evolueaza increderea clientilor in timp si identifica categoriile de produse „vedeta” ale magazinului online.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('online_store_data.csv')
df['date_added'] = pd.to_datetime(df['date_added'])
df['year'] = df['date_added'].dt.year
```

### PERSPECTIVE ANALITICE
Perspectiva 1: Identificarea valorilor extreme.
Calculam notele minime si maxime pentru a intelege plaja de evaluari a produselor. 

```python
rating_max = df['rating'].max()
rating_min = df['rating'].min()

print(f"Nota maxima primita de un produs: {rating_max}")
print(f"Nota minima primita de un produs: {rating_min}")
print("-" * 30)
```

### Perspectiva 2: Evolutia increderii in timp.
Analizam grafic media notelor pe fiecare an pentru a observa trendul satisfactiei clientilor.

```python
yearly_rating = df.groupby('year')['rating'].mean()

plt.figure(figsize=(10, 5))
plt.plot(yearly_rating.index, yearly_rating.values, marker='o', color='b', linestyle='-')
plt.title('Evolutia Notei Medii a Produselor pe Ani')
plt.xlabel('Anul adaugarii in oferta')
plt.ylabel('Nota Medie')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

### Perspectiva 3: Performanta pe categorii.
Comparam notele medii ale diferitelor categorii de produse.

```python
category_rating = df.groupby('category')['rating'].mean().sort_values(ascending=False)

plt.figure(figsize=(12, 6))
category_rating.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Top Categorii dupa Evaluarile Utilizatorilor')
plt.xlabel('Categorie Produs')
plt.ylabel('Nota Medie')
plt.xticks(rotation=45)
plt.show()
```

### Perspectiva 4: Analiza veniturilor.
Identificam categoriile care genereaza cele mai mari incasari.

```python
category_revenue = df.groupby('category')['revenue'].sum().sort_values(ascending=False)
print(category_revenue.head())
```

## CONCLUZII: 
**Observatii:**
* Satisfactia clientilor este in crestere, conform graficului de evolutie anuala.
* Categoriile **„Mice”** si **„Tablets”** sunt cele mai apreciate de utilizatori.

**Dificultati intampinate:**
* Cea mai mare provocare a fost fromatarea coloanei **'date_added'** pentru a permite gruparea pe ani.
