```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go

# Incarcam datele
df = pd.read_csv('online_store_data.csv')
```

```python
# Filtrare pentru cele 3 categorii cheie
categorii_cheie = ['Smartphones', 'Laptops', 'Tablets']
df_focus = df[df['category'].isin(categorii_cheie)].copy()
```

```python
# Ponderea veniturilor pe Brand (Pie Chart)
fig_pie = px.pie(df_focus, values='revenue', names='brand', 
             title='Ponderea Veniturilor in Funciie de Brand (Smartphones, Laptops, Tablets)',
             hole=0.3, 
             color_discrete_sequence=px.colors.sequential.RdBu)
fig_pie.update_traces(textinfo='percent+label')
fig_pie.show()
```

```python
# Vanzarile in functie de Brand (Line Chart)
# Grupam datele pentru a vedea totalul vanzarilor per brand
sales_brand = df_focus.groupby('brand')['quantity_sold'].sum().reset_index().sort_values(by='quantity_sold', ascending=False)

fig_line = px.line(sales_brand, x='brand', y='quantity_sold', 
              title='Volumul Vanzarilor per Brand',
              markers=True, text='quantity_sold')
fig_line.update_traces(textposition="top center")
fig_line.show()
```

```python
# Evaluarile Utilizatorilor (Histogram + Boxplot)
fig_dist = px.histogram(df_focus, x='rating', color='category', 
                   marginal='box',
                   title='Distributia Evaluarilor: Histograma si Boxplot pe Categorii',
                   hover_data=df_focus.columns)
fig_dist.show()
```

```python
# Pret vs. Vanzari (Scatter Plot)
fig_scatter = px.scatter(df_focus, x='price', y='quantity_sold', 
                    color='brand', size='rating', 
                    hover_name='product_name', 
                    title='Relatia Pret vs. Vanzari (Marimea punctului = Nota)',
                    log_x=True)
fig_scatter.show()
```

CONCLUZII:

DOMINANTA: Brandul Apple detine cea mai mare pondere din venituri si volumul de unitati vandute.

CALITATE: Tabletele au cele mai constante evaluari, grupate strans in jurul mediei, ceea ce indica o satisfactie uniforma. In schimb, la smartphone-uri observam o variatie mult mai mare; acest lucru sugereaza ca experienta utilizatorilor depinde critic de modelul ales, nu doar de brand.

STRATEGIA DE PRET: Datele confirma ca pretul ridicat nu reprezinta o bariera pentru vanzari, atata timp cat rating-ul produsului este peste 9.0. Clientii nostri sunt dispusi sa plateasca premium pentru calitate garantata.

VITEZA DE VANZARE: Branduri precum Samsung si Huawei mentin un volum ridicat de unitati vandute, fiind pilonii magazinului pentru fluxul de numerar constant.
