Obiectiv: Evaluarea tendintelor Netflix pentru a decide strategia de intrare a companiei noastre pe piata.

```python
# 1. Introducere si Obiectivele Analizei
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go

df = pd.read_csv('netflix_dataset.csv')

# Curatare rapida: transformam 'release_year' in intreg si gestionam valorile lipsa la scoruri
df['release_year'] = df['release_year'].fillna(0).astype(int)
df['imdb_score'] = df['imdb_score'].fillna(df['imdb_score'].mean())
```

```python
# 2. Evolutia continutului in timp
trend_data = df.groupby(['release_year', 'type']).size().reset_index(name='count')
# Filtram anii recenti pentru relevanta (ex: după 2000)
trend_data = trend_data[trend_data['release_year'] > 2000]

fig_trend = px.line(trend_data, x='release_year', y='count', color='type',
                  title='Evolutia Productiilor Netflix (2000-2021)',
                  labels={'count': 'Numar Productii', 'release_year': 'An Lansare'})
fig_trend.show()
```

```python
# 3. Analiza Calitatii 
df_clean = df.dropna(subset=['tmdb_popularity', 'tmdb_score'])

# Cream graficul scatter interactiv
fig_scatter = px.scatter(df_clean, 
                         x='tmdb_score', 
                         y='tmdb_popularity', 
                         color='type', 
                         size='runtime',
                         hover_name='title', 
                         title='Analiza Corelatiei: Scor TMDB vs Popularitate',
                         labels={'tmdb_score': 'Scor Utilizatori', 'tmdb_popularity': 'Popularitate'},
                         template='plotly_white')

fig_scatter.show()
```

```python
# 4. Distributia Genurilor si Preferintele Audientei
# Cream un grafic de tip Pie pentru a vedea proportia
fig_pie = px.pie(df, names='type', title='Ponderea Continutului: Filme vs. Seriale',
             color_discrete_sequence=px.colors.qualitative.Pastel)
fig_pie.update_traces(textinfo='percent+label')
fig_pie.show()
```

```python
# 5. Distributia Duratei Productiilor
# Histograma pentru durata continutului
fig_runtime = px.histogram(df, x='runtime', color='type', marginal='violin',
                       title='Distributia Duratei Continutului',
                       labels={'runtime': 'Durata (minute)'},
                       nbins=50)
fig_runtime.show()
```

CONCLUZII:
1. Datele arata o creștere accelerata a serialelor in ultimul deceniu. Pentru o platforma noua, serialele sunt esentiale pentru a mentine abonatii activi pe termen lung.
2. Am observat ca succesul comercial (popularitatea) nu necesita intotdeauna un scor critic de 10/10. Putem investi in continut "popcorn" care aduce audienta, chiar daca notele sunt medii.
3. Majoritatea filmelor de succes se incadreaza in intervalul 90-110 minute. Recomandam evitarea productiilor extrem de lungi pentru a maximiza numarul de vizionari complete.
